```python
def linear_regression(x,y):
    x=np.array(x).flatten()
    y=np.array(y).flatten()
    x_mean=x.mean()
    y_mean=y.mean()
    m=sum((x-x_mean)*(y-y_mean))/sum((x-x_mean)**2)
    m = sum((x-x_mean)*(y-y_mean))/sum((x-x_mean)**2)
    c = y_mean - (m * x_mean)
    return m,c
```

```python
for i in col:
    if ( i == 'CYLINDERS' ):
        continue
    slope, intercept =linear_regression(df[i], df['COEMISSIONS'])
    plt.title(f'LINEAR REGRESSION b/w {i} and COEMISSION')
    plt.xlabel(f'{i}')
    plt.ylabel('COEMISSION')
    plt.plot(df[i], slope*df[i]+intercept, color='black')
    plt.scatter(df[i], df['COEMISSIONS'], color='red', alpha=0.5)
    plt.show()
```

```python
def multi(df, features, target):
    X = df[features].values
    y = df[target].values.reshape(-1, 1)
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    print(f"Intercept (c): {theta[0][0]:.4f}")
    for name, value in zip(features, theta[1:].flatten()):
        print(f"Slope for {name}: {value:.4f}")

    y_pred = X_b.dot(theta)
    plt.figure(figsize=(12, 6))
    plt.title('Multiple Linear Regression')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.scatter(df[target], y_pred, color='black', alpha=0.5)
    plt.show()

    return theta, y_pred
```

```python
features = ['ENGINE SIZE','CYLINDERS','FUEL CONSUMPTION']
target = ['COEMISSIONS']
multi(df,features, target)
```
