```python
import pandas as pd
```

```python
df=pd.read_csv("DataClean.csv")
df
```

```python
text_cols = ['Status', 'Education', 'Industry', 'Location', 'AI Risk']
for col in text_cols:
    df[col] = df[col].str.lower().str.strip()
df
```

```python
df['Date Recorded'] = pd.to_datetime(df['Date Recorded'], errors='coerce')
df
```

```python
df['Location'].unique()
```

```python
df['Location'] = df['Location'].str.lower().str.replace('delhi','urban').str.replace('hyderabad','urban').str.replace('bangalore','urban').str.replace('mumbai','urban')
df
```

```python
df['Age Group'].unique()
```

```python
df['Age Group'] = df['Age Group'].str.replace('_','-').str.replace(' - ','-').str.replace('+','-65').str.replace(' to ','-')
df
```

```python
df['Education'] = df['Education'].str.lower()
df['Education'].unique()
```

```python
df['Education'] = df['Education'].str.replace('masters','master').str.replace('.','').str.replace('bachelors','bachelor').str.replace("'s","")
df
```

```python
df['Industry'].unique()
```

```python
df['Industry'] = df['Industry'].str.replace('technology','tech').str.replace('healthcare','health')
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Years of Experience','Date Recorded'], inplace=True)
df
```

```python
for col in ['Status','Age Group','Education','Industry','AI Risk','Location']:
    mode = df[col].mode()[0]
    df[col] = df[col].fillna(mode)
df
```

```python
df['Monthly Salary (INR)'] = df['Monthly Salary (INR)'].fillna(
    df.groupby(['Education', 'Industry', 'Location'])['Monthly Salary (INR)'].transform('mean')
)
df
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.reset_index(drop=True, inplace=True)
df
```
